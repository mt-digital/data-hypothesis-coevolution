```python
import matplotlib.pyplot as plt
import numpy as np
import warnings; warnings.simplefilter('ignore')

from experiment import Experiment

Ks = [b * (10**p) for b in range(1,11) for p in range(7)]
Ks.sort()

Ks_samp = Ks[1:-45:4]
# Ks_samp.remove(10)
print(Ks_samp)
# plt.plot(np.lg(Ks_samp))

sigma = 0.5
Ks = [5, 10, 50, 100, 200, 300, 400, 500, 750, 1000]
# print(Ks)
n_trials = 100
# ex = Experiment(Ks, [sigma], n_trials=n_trials)
ex = Experiment(Ks_samp, [sigma], n_trials=n_trials)
ex.run()
```

```python
average_its = [
    np.mean(
        [ex.trials[(k, sigma)][i]['d'].shape[0] for i in range(n_trials)]
    )
    for k in Ks_samp
]
```
